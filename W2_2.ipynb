```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import linalg
%matplotlib inline
```

Строим функцию $f(x)$

```python
def f(x):
    return math.sin(x / 5.) * math.e**(x / 10.) + 5. * math.e**(-x / 2.)
```

Нарежем аргументов для подстановки в функции в дальнейшем

```python
a = np.linspace(1, 15, 1000)
```

Функция $f(x)$ не может принять массив, поэтому делаем список с помощью map

```python
y = list(map(f,a))
```

Рисуем

```python
plt.plot(a,y);
```

Составляем матрицу А из значений 1 1 и 1 15, и вектор b из значений функции от 1 и 15

```python
A = np.array([[1.0, 1.0],[1.0, 15.0]])
b = np.array(list(map(f,[1., 15.])))
```

Решаем

```python
B1 = linalg.solve(A,b)
```

Составляем функцию подставляя на места коэффициентов найденные значения

```python
def g1(x):
    return B1[1] * x + B1[0]
```

Рисуем

```python
plt.plot(a, g1(a), a, y);
```

### Вторая степень многочлена

```python
A2 = np.array([[1., 1., 1.],[1., 8., 64.],[1.,15.,225.]])
b2 = np.array(list(map(f,[1, 8, 15])))
```

```python
B2 = linalg.solve(A2,b2)
```

```python
def g2(x):
    return (B2[2]*x**2. + B2[1]*x + B2[0])
plt.plot(a, g2(a), a, y);
```

### Третья степень

```python
A3 = np.array([[1., 1., 1., 1.],[1., 4., 16., 64. ],[1, 10, 100, 1000],[1.,15.,225., 3375]])
b3 = np.array(list(map(f,[1, 4, 10, 15])))
```

```python
B3 = linalg.solve(A3,b3)
```

```python
def g3(x):
    return (B3[3]*x**3 + B3[2]*x**2. + B3[1]*x + B3[0])
plt.plot(a, g3(a), a, y);
```
