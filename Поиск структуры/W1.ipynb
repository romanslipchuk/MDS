```python
import pandas as pd
from sklearn.cluster import MeanShift
```

```python
data = pd.read_table('./201309_foursquare_dataset_umn/fsq/umn_foursquare_datasets/checkins.dat', sep= '|', skipinitialspace=True, skiprows=[1])
```

```python
data.dropna(inplace=True)
data.columns = data.columns.str.strip()
data.set_index('id')
data.sort_index()
```

```python
X  = data[['latitude','longitude']][:100001]
```

```python
ms = MeanShift(bandwidth=0.1)
ms.fit(X)
```

```python
ms.cluster_centers_
```

```python
#new DataFrame from results
data_clustered = pd.DataFrame({"latitude": X.values[:,0],'longitude':X.values[:,1], "label": ms.labels_})
data_clustered.head()
```

```python
#Group by counts to sort > 15 checkins
data_sorted = data_clustered.groupby('label').count()
```

```python
clusters = data_sorted[data_sorted.latitude > 15].reset_index().label.values
```

```python
centers = []
for i in clusters:
    centers.append(ms.cluster_centers_[i])
```

```python
import numpy as np
centers_np = np.array(centers)
centers_np
```

```python
#Create offices array
offices = np.array([[33.751277, -118.188740],
                    [25.867736, -80.324116],
                    [51.503016, -0.075479],
                    [52.378894, 4.885084],
                    [39.366487, 117.036146],
                    [-33.868457, 151.205134]])
```

```python
#Calculate distance
def distance(point1, point2):
    return ((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)**0.5
```

```python

answer_index = 0
min_dist = 0
i = 0
while i < len(centers_np):
    distances = [distance(xx, centers_np[i]) for xx in offices]
    if min_dist == 0:
        min_dist = min(distances)
        answer_index = i
    else:
        if min_dist > min(distances):
            min_dist = min(distances)
            answer_index = i
    i += 1
```
