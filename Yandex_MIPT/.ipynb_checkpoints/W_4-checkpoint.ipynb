```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

### Экспоненциальное распределение

https://ru.wikipedia.org/wiki/%D0%AD%D0%BA%D1%81%D0%BF%D0%BE%D0%BD%D0%B5%D0%BD%D1%86%D0%B8%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5_%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5

Функция плотности распределения: $f(x) = λe^{-λx}$

Среднее: $μ=1/λ$

Дисперсия: $σ^2=1/λ^2$

https://docs.scipy.org/doc/scipy-1.2.1/reference/generated/scipy.stats.expon.html#scipy.stats.expon

```python
# зададим лямбду как 0.5 а scale = 1/lambda = 2 чтобы не совсем просто все было
rex = sts.expon(scale=2)
```

```python
# строим выборку на 1000 элементов
sample = rex.rvs(1000)

# создаем выборку из 1000 элементов на интервале от 0 до 10
x = np.linspace(0,10,100)

# строим гистограмму
pdf = rex.pdf(x)
plt.hist(sample, density=True, label='histogram')

# строим график теоретической плотности распределения
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# Строим распределение выборочного среднего случайной величины для выборок объёма 5, 10, 50:
for n in [5, 10, 50]:
    sample_means = []
    for rvn in range(1000):
        sample_means.append(sum([i/n for i in rex.rvs(n)])) #не пользуемся .mean()
        
    # и строим гистграмму
    plt.hist(sample_means, 100, density=True, range=[0, 12], label='sample means histogram')
    
    # вычисляем pdf плотность нормального распределения с данными параметрами 
    x = np.linspace(0, 4, 100)
    loc = 2 # среднее равно scale
    scale_norm = np.sqrt(loc**2 / n) # стандартное отклонение
    norm_rv = sts.norm.pdf(x ,loc , scale_norm)
    
    # рисуем
    plt.plot(x, norm_rv, label='normal theoretical pdf')
    plt.legend()
    plt.title("Sample size {}".format(n))
    plt.ylabel('fraction of samples $f(x)$')
    plt.xlabel('sample mean $x$')
    plt.show()
```

при возрастании объема выборки распределение выборочных средних всё лучше и лучшее описывается нормальным распределением. ЦПТ работает так как нужно!
